```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
from scipy import integrate, stats
import scipy
import time
import warnings
import copy
import pandas as pd
import seaborn as sns
from prettytable import PrettyTable
from numpy import random
import random as rnd
import math
from tqdm import tqdm 
import tabulate
from numba import njit, jit

warnings.filterwarnings("ignore", category=RuntimeWarning)
K = 99 # strike price
S0 = 100 # current price of the stock
N = 50 # time steps for each path
sigma = 0.2 # volatility
r = 0.06 # risk free rate 
T = 1 # year
```

```python
@jit(nopython=True)
def binom(K, S0, N, sigma, r, T, optiontype='put'):
    assert optiontype in ['call','put'], 'optiontype should be either "call" or "put".'
    dt = T/N
    u = np.exp(sigma*np.sqrt(dt))
    d = 1/u
    p = (np.exp(r*dt) - d) / (u - d)
    
    option_tree = {}
    option_tree[N] = []
    for j in range(N, -1, -1):
        price = S0*(u**j)*(d**(N-j))
        if optiontype == 'call':
            option_tree[N].append( max(price-K, 0) )
        elif optiontype == 'put':
            option_tree[N].append( max(K-price, 0) )
    
    for i in range(N-1, -1, -1):
        option_tree[i] = []
        for j in range(i+1):
            opt_price = np.exp(-r*dt)*(p*option_tree[i+1][j] + (1-p)*option_tree[i+1][j+1])
            option_tree[i].append( max(opt_price, 0) )
    
    return option_tree[0][0]

@jit(nopython=True)
def blackschol(St, K, sigma, r, T, t, optiontype='put'):
    assert optiontype in ['call','put'], 'optiontype should be either "call" or "put".'
    d1 = (1 / (sigma * np.sqrt(T-t)))*( np.log(St/K) + (r + (sigma**2/2))*(T-t))
    d2 = (1 / (sigma * np.sqrt(T-t)))*( np.log(St/K) + (r - (sigma**2/2))*(T-t))
    
    if optiontype == 'call':
        value = norm.cdf(d1)* St - norm.cdf(d2)* K * np.exp(-r *(T-t))
    elif optiontype == 'put':
        value = norm.cdf(-1*d2) * K * np.exp(-r *(T-t))  - norm.cdf(-1*d1)* St 
    
    return value

@jit(nopython=True)
def GBM(K, S0, sigma, r, T):
    Z = random.normal(0,1)
    S_T = S0*np.exp((r-0.5*sigma**2)*T + sigma*np.sqrt(T)*Z)
    return S_T

@jit(nopython=True)
def hedge_analytical(St, K, sigma, r, T, t, optiontype='put', digital=False):
    assert optiontype in ['call','put'], 'optiontype should be either "call" or "put".'
    if digital == False:
        d1 = (1 / (sigma * np.sqrt(T-t)))*( np.log(St/K) + (r + (sigma**2/2))*(T-t))
        if optiontype == 'call':
            return norm.cdf(d1)
        elif optiontype == 'put':
            return norm.cdf(d1) - 1
    else:
        d2 = (1 / (sigma * np.sqrt(T-t)))*( np.log(St/K) + (r - (sigma**2/2))*(T-t))
        if optiontype == 'call':
            return norm.pdf(d2)*np.exp(-r*(T-t))*(1/(sigma*St*np.sqrt((T-t))))
        elif optiontype == 'put':
            return (-1)*norm.pdf(d2)*np.exp(-r*(T-t))*(1/(sigma*St*np.sqrt((T-t))))

@jit(nopython=True)
def Payoff(K, S0, N, sigma, r, T, optiontype = 'put',digital=False):
    assert optiontype in ['call','put'], 'optiontype should be either "call" or "put".'
    S_N = GBM(K, S0, sigma, r, T)
    if digital == False:
        if optiontype=='put':
            payment = max(K - S_N,0)
        elif optiontype=='call':
            payment = max(S_N - K,0)
    else:
        if optiontype == 'call':
            if K<S_N:   
                payment = 1
            else:
                payment = 0
        elif optiontype == 'put':
            if K>S_N:
                payment = 1
            else:
                payment = 0
                
    return payment

@jit(nopython=True)
def MC_DAVPO(K, S0, N, sigma, r, T, M, optiontype='put'):
    assert optiontype in ['call','put'], 'optiontype should be either "call" or "put".'
    payoff_list = np.zeros(M)

    for i in range(M):
        payoff_list[i] = Payoff(K, S0, N, sigma, r, T, optiontype ,digital=False)

    disc_payoff = [np.exp(-1*r*T)*i for i in payoff_list]
    #confidence_level = 0.95
    #degrees_freedom = M - 1
    sample_mean = np.mean(disc_payoff)
    sample_standard_error = scipy.stats.sem(disc_payoff)
    #confidence_interval = scipy.stats.t.interval(confidence_level, degrees_freedom, sample_mean, sample_standard_error)
    
    return sample_mean, sample_standard_error 
    
@jit(nopython=True)
def MC_digi(K, S0, N, sigma, r, T, M):
    assert optiontype in ['call','put'], 'optiontype should be either "call" or "put".'
    payoff = np.zeros(M)
    for i in range(M):
        S_N = GBM(K, S0, sigma, r, T)
        if K<S_N:   
            payoff[i] = 1
        else:
            payoff[i] = 0

    V = math.exp(-1*r*T)*np.mean(payoff)
    std_payoff = np.std(payoff)
    std_MC = std_payoff/math.sqrt(M)
    return V,std_MC
```

### MC simulation

```python
def MC_simulation(S0=100, r=0.06,sigma=0.2,N =50,T=1, nSims=100):

    dt = T/N
    mu = (r-0.5*sigma**2)*dt
    sigma = sigma*np.sqrt(dt)
    St = np.zeros(shape=(N,nSims))
    St[0,] = S0
    for i in range(1,N):
            for j in range(0,nSims):
                e = random.normal(0,1)
                St[i,j] = St[i-1,j]*np.exp(mu+sigma*e)
    return St
plt.plot(pd.DataFrame(MC_simulation()).head())
plt.title("Monte Carlo Simulations")
plt.xlabel("Number of Simulations")
plt.ylabel("Simulated Stock Prices")
plt.show()
```

### Digital option - sophisticated methods

```python
#Likelihood Ratio method
LR_result = []
M=1000

def LR_payoff(K, S0, sigma, r, T):
    Z = random.normal(0,1)
    S_T = S0*math.exp((r-0.5*sigma**2)*T + sigma*math.sqrt(T)*Z)
    if K<S_T:   
        payment = 1
    else:
        payment = 0
    return payment,Z

for i in range(M):
    I,Z = LR_payoff(K, S0, sigma, r, T)
    LR_result.append(math.exp(-1*r*T)*I*Z/(S0*sigma*math.sqrt(T)))

print("The mean and variance of delta:")
print(np.mean(LR_result),np.var(LR_result))
```

```python
def LR_payoff(K, S0, sigma, r, T, optiontype = 'call', digital = False):
    Z = random.normal(0,1)
    S_T = S0*np.exp((r-0.5*sigma**2)*T + sigma*np.sqrt(T)*Z)
    if digital == False:
        if optiontype == 'call':
            payment = max(S_T - K, 0)
        elif optiontype == 'put':
            payment = max(K - S_T, 0)
            
    elif digital == True:
        if optiontype == 'call':
            if K<S_T:   
                payment = 1
            else:
                payment = 0
        elif optiontype == 'put':
            if K>S_T:
                payment = 1
            else:
                payment = 0
                
    return payment,Z

t4 = PrettyTable(["Sample size", "Absolute error", "Relative error"])
t5 = PrettyTable(["Sample size", "Absolute error", "Relative error"])
delta_BS = hedge_analytical(S0, K, sigma, r, T, 0,'put',False)

for j in tqdm(M): 
    LR_result_van = []
    LR_result_digi = []
    for k  in range(j):
        I,Z = LR_payoff(K, S0, sigma, r, T, 'put', False)
        LR_result_van.append(np.exp(-1*r*T)*I*Z/(S0*sigma*np.sqrt(T)))
        I1,Z1 = LR_payoff(K, S0, sigma, r, T, 'call', True)
        LR_result_digi.append(np.exp(-1*r*T)*I1*Z1/(S0*sigma*np.sqrt(T)))

    delta_van = np.mean(LR_result_van)
    delta_digi = np.mean(LR_result_digi)

    t4.add_row([j, "{:.5f}".format(np.abs((delta_van-delta_BS))), "{:.0%}".format(np.abs((delta_van-delta_BS)/delta_BS))])
    t5.add_row([j, "{:.5f}".format(np.abs((delta_digi-delta_BS_digi))), "{:.0%}".format(np.abs((delta_digi-delta_BS_digi)/delta_BS_digi))])
    

print(t4)
print(t5)
```

```python
# Numerical Integration using Scipy integarte
eps=0.2
func1 = lambda S: (np.exp(-1*r*T)/(2*eps*S0*np.sqrt(2*T*np.pi*sigma**2)))*(1-(np.tanh((S-K)/eps))**2)*np.exp((-(np.log(S/S0)-(r-(0.5*sigma**2))*T)**2)/(2*T*sigma**2))

print("Delta using Scipy's numerical integration: ", integrate.quad(func1, 85, np.inf)[0])
```

```python
@jit(nopython=True)
def smooth_func(S, eps=0.2, sigma=0.2, r=0.06, T=1, S0=100, K=99):
    return (np.exp(-1*r*T)/(2*eps*S0*np.sqrt(2*T*np.pi*sigma**2)))*(1-(np.tanh((S-K)/eps))**2)*np.exp((-(np.log(S/S0)-(r-(0.5*sigma**2))*T)**2)/(2*T*sigma**2))

print("Delta using Scipy's numerical integration (eps=0.2): ", integrate.quad(smooth_func, 85, np.inf)[0])

@jit(nopython=True)
def smooth_func2(S, eps=0.2, sigma=0.2, r=0.06, T=1, S0=100, K=99):
    return (1/S**2)*(np.exp(-1*r*T)/(2*eps*S0*np.sqrt(2*T*np.pi*sigma**2)))*(1-(np.tanh(((1/S)-K)/eps))**2)*np.exp((-(np.log(1/(S*S0))-(r-(0.5*sigma**2))*T)**2)/(2*T*sigma**2))
    
```

```python
# func2 = lambda S: (1/S**2)*(np.exp(-1*r*T)/(2*eps*S0*np.sqrt(2*T*np.pi*sigma**2)))*(1-(np.tanh(((1/S)-K)/eps))**2)*np.exp((-(np.log(1/(S*S0))-(r-(0.5*sigma**2))*T)**2)/(2*T*sigma**2))
@jit(nopython=True)
def MC_integrate(num_samples=10**5, epsilon=0.2):
   
    sum_of_samples1 = 0
    sum_of_samples2 =0
    for i in range(num_samples):
        x = np.random.uniform(0,1)
        sum_of_samples1 += smooth_func(S=x, eps=epsilon)
        
    for i in range(num_samples):
        x = np.random.uniform(0,1)
        sum_of_samples2 += smooth_func2(S=x, eps=epsilon)  
     
    I1 =  (sum_of_samples1/num_samples)
    I2 =  (sum_of_samples2/num_samples)
    return I1 + I2

print("Delta using Monte carlo integration: ", MC_integrate(10**6))
```

```python
epslist = np.linspace(-100,50,500)
numerical = []
montecarlo = []

for i in tqdm(epslist):
    numerical.append(integrate.quad(smooth_func, 85, np.inf, args=(i,))[0])
    montecarlo.append(MC_integrate(10**6, epsilon=i))
```

### Asian option

```python
@jit(nopython=True)
def Analytical_Geo_AsianCall(S0=100, K=99, sigma=0.2, r=0.06, T=1, N=50):
    sigma_new = sigma*np.sqrt((2*N+1)/(6*N+6))
    r_new = ((r-0.5*sigma**2)+sigma_new**2)/2

    d1_new = (np.log(S0/K) + (r_new + 0.5*sigma_new**2)*T)/(sigma_new*np.sqrt(T))
    d2_new = (np.log(S0/K) + (r_new - 0.5*sigma_new**2)*T)/(sigma_new*np.sqrt(T))
    
    value = np.exp(-1*r*T)*(S0*np.exp(r_new*T)*norm.cdf(d1_new) - K*norm.cdf(d2_new))
    return value
```

```python
Asian_analytic = Analytical_Geo_AsianCall()
print("The analytical price of an Asian call option based on geometric average: ", "{:.3f}".format(Asian_analytic))
```

```python
@jit(nopython=True)
def GBM_path(S0=100, K=99, sigma=0.2, r=0.06, T=1, N=50):
    dt = T/N
    St = np.zeros(N)
    St[0] = S0
    mu = (r-(0.5*sigma**2))*dt
    sigma = sigma*np.sqrt(dt)
    for i in range(1, N):
        e = random.normal(0,1)
        St[i] = St[i-1]*np.exp(mu+sigma*e)
    
    return St

@jit(nopython=True)
def Arith_payoff(path,K=99,r=0.06,T=1):
    payoff = max(np.mean(path)-K,0)*np.exp(-1*r*T)
    return payoff

@jit(nopython=True)
def Geo_payoff(path, K=99,r=0.06,T=1):
    payoff = max(np.exp(np.mean(np.log(path)))-K, 0)*np.exp(-1*r*T)
    return payoff

@jit(nopython=True)
def MC_Geo_Asian(K=99, S0=100, N=50, sigma=0.2, r=0.06, T=1, M=10**4):
    geo_pay = np.zeros(M)
    for i in range(M):
        path = GBM_path(S0, K, sigma, r, T, N)
        geo_pay[i] = Geo_payoff(path, K,r,T)
    
    mean = np.mean(geo_pay)
    error = np.std(geo_pay)/np.sqrt(M)
    
    return mean , error
    
@jit(nopython=True)
def arith_AsianCall_MC_withoutCV(S0=100, sigma=0.2, K=99, r=0.06, T=1, N=50, M = 10**4):
    arith_payoff = np.zeros(M)
    for i in range(M):
        path = GBM_path(S0, K, sigma, r, T, N)
        arith_payoff[i] = Arith_payoff(path,K,r,T)

    mean = np.mean(arith_payoff)
    var = np.var(arith_payoff)
    #error = scipy.stats.sem(arith_payoff)
    #arithMC  = np.mean(arith_payoff)
    #arithMC_error = np.std(arith_payoff)/np.sqrt(M)
    
    return mean, var

@jit(nopython=True)
def arith_Asiancall_MC_CV(S0=100, sigma=0.2, K=99, r=0.06, T=1, N=50, M = 10**4): #Estimates expected payoff from arithmetic Asian with controlvariate
    geo = np.zeros(M)
    ari = np.zeros(M)
    for n in range(M):
        path = GBM_path(S0, K, sigma, r, T, N)
        ari[n]= Arith_payoff(path, K,r,T)
        geo[n]= Geo_payoff(path, K,r,T)
        
    S_xy=0
    Var_x=0
    for n in range (M):
        S_xy += (geo[n] - np.mean(geo)) * (ari[n] - np.mean(ari))
        Var_x += (geo[n] - np.mean(geo))**2
        
    bstar = S_xy/Var_x
        
    result = np.mean(ari) - bstar*(np.mean(geo)- Analytical_Geo_AsianCall(S0, K, sigma, r, T, N))
    var = (1-np.corrcoef(ari,geo)**2)*np.var(ari)
    
    return result, var
```

```python
# Control variate technique 

K_vals = np.arange(50,121,1); withcv_K = []; nocv_K = []; withcv_K_var = []; nocv_K_var = []
for i in tqdm(K_vals):
    temp_list1 = []
    temp_list1var = []
    temp_list2 = []
    temp_list2var = []
    for j in range(30):
        mu1, var1 = arith_AsianCall_MC_withoutCV(K = i)
        temp_list1.append(mu1)
        temp_list1var.append(var1)
        mu2, var2 = arith_Asiancall_MC_CV(K = i)
        temp_list2.append(mu2)
        temp_list2var.append(var2) 
    nocv_K.append(np.mean(temp_list1))
    nocv_K_var.append(np.mean(temp_list1var))
    withcv_K.append(np.mean(temp_list2))
    withcv_K_var.append(np.mean(temp_list2var))
                          
N_vals = np.arange(5,500,5); withcv_N = []; nocv_N = []; withcv_N_var = []; nocv_N_var = [] 
for i in tqdm(N_vals):
    temp_list1 = []
    temp_list1var = []
    temp_list2 = []
    temp_list2var = []
    for j in range(30):  
        mu1, var1 = arith_AsianCall_MC_withoutCV(N = i)
        temp_list1.append(mu1)
        temp_list1var.append(var1)
        mu2, var2 = arith_Asiancall_MC_CV(N = i)
        temp_list2.append(mu2)
        temp_list2var.append(var2)
    nocv_N.append(np.mean(temp_list1))
    nocv_N_var.append(np.mean(temp_list1var))
    withcv_N.append(np.mean(temp_list2))
    withcv_N_var.append(np.mean(temp_list2var))
    
sim_vals = np.arange(10**2,3*10**4,500); withcv_sim = []; nocv_sim = []; withcv_sim_var = []; nocv_sim_var = []
for i in tqdm(sim_vals):
    temp_list1 = []
    temp_list1var = []
    temp_list2 = []
    temp_list2var = []
    for j in range(30):
        mu1, var1 = arith_AsianCall_MC_withoutCV(M = i)
        temp_list1.append(mu1)
        temp_list1var.append(var1)
        mu2, var2 = arith_Asiancall_MC_CV(M = i)
        temp_list2.append(mu2)
        temp_list2var.append(var2)
    nocv_sim.append(np.mean(temp_list1))
    nocv_sim_var.append(np.mean(temp_list1var))
    withcv_sim.append(np.mean(temp_list2))
    withcv_sim_var.append(np.mean(temp_list2var))
    
sigma_vals = np.arange(0,1.5,0.05); withcv_sigma = []; nocv_sigma = []; withcv_sigma_var = []; nocv_sigma_var = [] 
for i in tqdm(sigma_vals):
    temp_list1 = []
    temp_list1var = []
    temp_list2 = []
    temp_list2var = []
    for j in range(30):
        mu1, var1 = arith_AsianCall_MC_withoutCV(sigma = i)
        temp_list1.append(mu1)
        temp_list1var.append(var1)
        mu2, var2 = arith_Asiancall_MC_CV(sigma = i)
        temp_list2.append(mu2)
        temp_list2var.append(var2)
    nocv_sigma.append(np.mean(temp_list1))
    nocv_sigma_var.append(np.mean(temp_list1var))
    withcv_sigma.append(np.mean(temp_list2))
    withcv_sigma_var.append(np.mean(temp_list2var))
    
r_vals = np.arange(0,1.5,0.01); withcv_r = []; nocv_r = []; withcv_r_var = []; nocv_r_var = []
for i in tqdm(r_vals):
    temp_list1 = []
    temp_list1var = []
    temp_list2 = []
    temp_list2var = []
    for j in range(30):
        mu1, var1 = arith_AsianCall_MC_withoutCV(r = i)
        temp_list1.append(mu1)
        temp_list1var.append(var1)
        mu2, var2 = arith_Asiancall_MC_CV(r = i)
        temp_list2.append(mu2)
        temp_list2var.append(var2)
    nocv_r.append(np.mean(temp_list1))
    nocv_r_var.append(np.mean(temp_list1var))
    withcv_r.append(np.mean(temp_list2))
    withcv_r_var.append(np.mean(temp_list2var))

T_vals = np.arange(1,10,0.25); withcv_T = []; nocv_T = []; withcv_T_var = []; nocv_T_var = []
for i in tqdm(T_vals):
    temp_list1 = []
    temp_list1var = []
    temp_list2 = []
    temp_list2var = []
    for j in range(30):
        mu1, var1 = arith_AsianCall_MC_withoutCV(T = i)
        temp_list1.append(mu1)
        temp_list1var.append(var1)
        mu2, var2 = arith_Asiancall_MC_CV(T = i)
        temp_list2.append(mu2)
        temp_list2var.append(var2)
    nocv_T.append(np.mean(temp_list1))
    nocv_T_var.append(np.mean(temp_list1var))
    withcv_T.append(np.mean(temp_list2))
    withcv_T_var.append(np.mean(temp_list2var))
```
